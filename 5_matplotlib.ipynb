```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# одномерный массив
# x_arr = np.arange(-10, 11)
x_arr = np.arange(10)
x_arr
```

```python
# функция
y_arr_1 = x_arr ** 2
y_arr_2 = x_arr ** 4
```

```python
# визуализация 
plt.figure(figsize=(5, 5))
plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)
plt.show()
```

```python
# одномерный массив из 100 знаяений
# линейная последовательность в диапазоне от -10 до 10
x_arr = np.linspace(-10, 10, 100)
x_arr
```

```python
# функция синуса
y_1 = np.sin(x_arr)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr)

# функции сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_3 = sigmoid(x_arr)

```

```python
# по названию цвета
plt.plot(x_arr, y_1, color="red")
# по первой букве названия цвета из форматов RGB или CMYP
plt.plot(x_arr, y_2, color="y")
# по 16 кодировке RGB или RGBA
plt.plot(x_arr, y_3, c="#f9ff20")

plt.show()
```

```python
# типы линий

y_1 = np.sin(x_arr)
y_2 = np.sin(x_arr + 1)
y_4 = np.sin(x_arr + 2)

# штриховая линия
plt.plot(x_arr, y_1, linestyle="dashed")
#  пунктирная
plt.plot(x_arr, y_2, linestyle=":")
# штрих пунктирная
plt.plot(x_arr, y_4, linestyle="-.", c="#ff4500")

plt.show()
```

```python
# объединение настроек цвета и типа линий

plt.plot(x_arr, y_1, "--g", lw=2)
plt.plot(x_arr, y_2, ":m")
plt.plot(x_arr, y_4, "+r", lw=1)
plt.show()
```

```python
# заголовок, легенда графика, подпичей осей

# размер окна
plt.figure(figsize=(10, 7))

plt.plot(x_arr, y_1, ":m")
plt.plot(x_arr, np.cos(x_arr), "+r")

plt.title("График синуса и косинуса", c="r")

# легенда графика 
plt.legend(["синус", "косинус"])

# подпись осей
plt.xlabel("ось х")
plt.ylabel("ось у")


plt.show()
```
